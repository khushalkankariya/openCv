```python
#BGR to GRAY
img=cv.imread('D:/openCV/photos/chutiya.jpg')
gray=cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow("gray",gray)
cv.imshow("original",img)
cv.waitKey(0)
```

```python
#BGR to HSV
img=cv.imread('D:/openCV/photos/chutiya.jpg')
hsv=cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv.imshow("HSV",hsv)
cv.imshow("original",img)
cv.waitKey(0)
```

```python
#BGR to L*A*B
#BGR to HST
img=cv.imread('D:/openCV/photos/chutiya.jpg')
LAB=cv.cvtColor(img, cv.COLOR_BGR2LAB)
cv.imshow("LAB",LAB)
cv.imshow("original",img)
cv.waitKey(0)
```

```python
# L*: Lightness
# a*: Red/Green Value
# b*: Blue/Yellow Value
```

```python
#opencv reads images in BGR
#matplotlib reads as RGB image 
# so you will se inversion of colors
import matplotlib.pyplot as plt
img=cv.imread('D:/openCV/photos/chutiya.jpg')
cv.imshow("original",img)
cv.waitKey(0)


plt.imshow(img)
```

```python
#BGR To RGB
import matplotlib.pyplot as plt
%matplotlib inline 
img=cv.imread('D:/openCV/photos/darling.jpg')
RGB=cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.imshow("RGB_open_cv",RGB)
#Notice the color difference
#TO display proper image in matplotlib pass RGB img whereas in open cv pass BGR img
%matplotlib qt
plt.imshow(RGB)

plt.show()
cv.waitKey(0) 
```

```python
# we can convert all above image back to BGR but if you want to convert ex- HSV to GRAYSCALE the BGR will be checkpoint.
#HSC to BGR
img=cv.imread('D:/openCV/photos/darling.jpg')
hsv=cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv.imshow("HSV",hsv)
hsv_to_bgr=cv.cvtColor(hsv, cv.COLOR_HSV2BGR)
cv.imshow("HSV_TO_BGR",hsv_to_bgr)
cv.waitKey(0) 
```
